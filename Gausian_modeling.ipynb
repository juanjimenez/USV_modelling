Ejemplo de random process para ilustrar el kriging
Vamos a tomar una aproximaciòn bayesiana.
$p(x|y) = \mathcal{N}(\mu_x+\Sigma_{xy}Sigma_y^{-1}(y-\mu_y),\Sigma_x-\Sigma_{xy}\Sigma_y^{-1}\Sigma_{xy}^T)$

Empleamos un kernel tambien gausiano para obtener las varianza
 
$m(x) = 0$

$k(x,x')=\theta_1\exp(-\frac{\theta_2}{2}(x-x')^2)$


```python
import numpy as np
def exponential_cov(x,y,params):
    return params[0]*np.exp(-0.5*params[1]*np.subtract.outer(x,y)**2)

def condicional(x_new,x,y,params):
    B = exponential_cov(x_new,x,params)
    C = exponential_cov(x,x,params)
    A = exponential_cov(x_new,x_new,params)
    Ain = np.linalg.inv(C)
    mu = B@Ain@y
    sigma = A-B@Ain@B.T
    return (mu.squeeze(),sigma.squeeze())




                            
```

```python
params = [1,1]
x = np.array([1,2,3])
y = np.array([2,-1,0])
salida = exponential_cov(x,y,params)
print(salida)
```

Ahora definimos un función para estimar valores a partir de los que ya conocemos


```python
def predict(x,data,kernel,params,sigma,t):
    k = [kernel(x,y,params) for y in data]
    Sinv = np.linalg.inv(sigma)
    y_pred = k@Sinv@t
    Sigma_new = kernel(x,x,params)- k@Sinv@k
    return y_pred, sigma_new
    
```
